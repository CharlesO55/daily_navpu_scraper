```python
import requests
from bs4 import BeautifulSoup
from datetime import datetime
import csv
```

```python
OUTPUT_FILE = "../data/chinabank.csv"

URL = 'https://www.chinabank.ph/forex-and-navpu'
response = requests.get(URL)

assert response.status_code == 200

soup = BeautifulSoup(response.content)
```

```python
uitf_table = soup.find_all('table')[1]

dt = uitf_table.find(class_=  "as-of-date").text
dt = datetime.strptime(dt, "As of %I:%M %p, %m/%d/%Y")
print(dt)


output = []
for tr in uitf_table.find_all('tr')[2:]:
    fund_name, fund_value = [t.text.strip() for t in tr.find_all('td')]
    
    print(fund_name, fund_value)
    output.append([dt, fund_name, fund_value])

with open(OUTPUT_FILE, 'a', newline='') as f:
    writer = csv.writer(f)
    writer.writerows(output)
```
